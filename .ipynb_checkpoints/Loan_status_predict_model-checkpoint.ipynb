```python
# Divide Train and Test data
```

```python
x = dataset.iloc[:,:-1]
y = dataset['loan_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=7)
```

```python
x.shape, x_train.shape, x_test.shape, y_train.shape
```

```python
y_test.value_counts()
```

```python
sc = StandardScaler()
sc.fit(x_train)
```

```python
x_train = pd.DataFrame(sc.transform(x_train),columns=x_train.columns)
```

```python
x_test = pd.DataFrame(sc.transform(x_test),columns=x_test.columns)
```

```python
# Model Train
```

```python
lr = LogisticRegression(max_iter=100)
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
# lr.get_params()
```

```python
mse = mean_squared_error(y_test,lr.predict(x_test))
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,lr.predict(x_test))
print(f"mean_absolute_error {mae}")
print(f"mean_squared_error {mse}")
print(f"root_mean_squar_error {rmse}")
```

```python
confusion_matrix(y_test,lr.predict(x_test))
```

```python
print("Accuracy result", classification_report(y_test,lr.predict(x_test)))
```

```python
sm = SVC()
sm.fit(x_train,y_train)
sm.score(x_train,y_train)*100, sm.score(x_test,y_test)*100
```

```python
mse1 = mean_squared_error(y_test,sm.predict(x_test))
rmse1 = np.sqrt(mse1)
mae1 = mean_absolute_error(y_test,sm.predict(x_test))
print(f"mean_absolute_error {mae1}")
print(f"mean_squared_error {mse1}")
print(f"root_mean_squar_error {rmse1}")
```

```python
confusion_matrix(y_test,sm.predict(x_test))
```

```python
print("Accuracy result", classification_report(y_test,sm.predict(x_test)))
```

```python
models = [
    ("logistic_Regrassion", lr),
    ("Super_vector_machin", sm),
    ("Decision_tree", dt),
    ("Random_forest", rm),
    ("GussianNB_model", gnb),
    ("knn_model", km)
]
```

```python
results = []
for name, model in models:
    
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    y_prob = model.predict_proba(x_test)[:,1] if hasattr(model, "predict_proba") else None
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "MSE" :  mean_squared_error(y_test,y_pred),
        
        "ROC-AUC": roc_auc_score(y_test, y_prob, multi_class='ovr') if y_prob is not None else None
    })
```

```python

classification_model_df = pd.DataFrame(results)
print(classification_model_df)
```
